```python
import itertools as it
import torch
import torch.nn as nn
from torch import Tensor
from typing import Tuple



# def vector_repr_2d(x_c,y_c):
#     assert len(x_c.size())==4
#     nb,nseqlen,nchxndim = x_c.size()
#     out,out2 =[],[]
#     for j in range(x_c.size(dim=-1)):
#         tmp_ch = x_c[:,:,j].unsqueeze(dim=-1)
#         tmp_ch = torch.cat([tmp_ch,(j*torch.ones_like(tmp_ch)).to(tmp_ch.device)],dim=-1)
#         out.append(tmp_ch)
        
#         tmp_ch2 = y_c[:,:,j].unsqueeze(dim=-1)
#         out2.append(tmp_ch2)
#     out = torch.cat(out,dim=-2)
#     out2 = torch.cat(out2,dim=-2)    
#     assert out.size(-1) == 2

#     out_pos,out_ch = out.split((out.size(-1)-1,1),dim=-1)
#     out_val = out2 
#     out_bool = torch.ones_like(out_pos).to(out_pos.device).bool()    
#     return out_pos,out_ch.int()[...,0],out_val,out_bool



def index_points(points, idx):
    """Get pointsï¼»idx]
    Args:
        points: input point-cloud data, (B, N, D)
        idx: sampled data index, (B, S)
    Returns:
        #new_points: indexed_points data, (B, S, D)
        new_points: indexed_points data, (B, S, nbd, D)
        
    """
    device = points.device
    B = points.size(0)
    index_shape = list(idx.shape)  # [B, S]
    index_shape[1:] = [1] * (len(index_shape) - 1)  # [B, 1]
    repeat_shape = list(idx.shape)
    repeat_shape[0] = 1
    batch_indices = torch.arange(B, device=device, dtype=torch.long).reshape(index_shape).repeat(repeat_shape)
    new_points = points[batch_indices, idx, ...]
    
    return new_points



class Distance_multichannels(nn.Module):
    def __init__(self,nchannels=3):
        super(Distance_multichannels, self).__init__()                
        self.nchannels = nchannels
        self.comb_idx = list(it.product(range(self.nchannels),repeat=2))
        
                            
    #def compute_distance(self,(x_s,ch_s),(x_a,ch_a),param_l):
    def compute_distance(self,sample_pair,all_pair,param_l):
        x_s,ch_s = sample_pair
        x_a,ch_a = all_pair
        
        distance_all = self.compute_distance_allchannels(x_s,x_a,param_l)
        all_idx = self.compute_jointindex(ch_s,ch_a)
        selected_bool = self.compute_selectedbool(all_idx,self.comb_idx)
        distance = torch.where(selected_bool.to(distance_all.device),distance_all,torch.zeros_like(distance_all).to(distance_all.device)).sum(dim=-1)
        return distance 
    
    
    def compute_distance_allchannels(self,src,target,comb_paraml):
        """
        input:
            src: (nb,nseqlen1,ndim) 
            target: (nb,nseqlen2,ndim) 
            comb_paraml: ( #pair_channels),ndim) 
        output:
            distance: (nb,nseqlen1,nseqlen1,#comb_paraml)         
        """
        
        total_channels = len(self.comb_idx)
        B, N, _ = src.shape
        _, M, _ = target.shape

        comb_paraml_ = (comb_paraml.T)[None,None,...]
        src = (src[:,:,None,:]).repeat(1,1,1,total_channels)        #(nb,m,d,comb_nch)
        target = (target[:,None,:,:]).repeat(1,1,1,total_channels)  #(nb,n,d,comb_nch)
        src /= comb_paraml_
        target /= comb_paraml_

        dist = -2*torch.einsum('bijk,bjlk->bilk',src,target)
        dist += (src**2).sum(dim=-2,keepdim=True)
        dist += (target**2).sum(dim=1,keepdim=True)
        #return dist
        return 1.-torch.exp(-dist)
    

    def compute_jointindex(self,ch_s,ch_a):
        nb,nch_s = ch_s.size()
        nb,nch_a = ch_a.size()
        all_idx = np.array([list(it.product(i[0],i[1])) for i in list(zip(ch_s.numpy(),ch_a.numpy()))]).reshape(nb,nch_s,nch_a,-1)
        return all_idx


    def compute_selectedbool(self,all_idx,comb_idx):
        selected_bool = []
        for ij_idx in comb_idx:
            selected_bool.append((np.sum(all_idx == ij_idx,axis=-1)==2)[...,np.newaxis])
        selected_bool= np.concatenate(selected_bool,axis=-1)
        selected_bool=torch.from_numpy(selected_bool)
        return selected_bool



class Nbdsampler_multichannels(nn.Module):
    def __init__(self,nchannels=3, nnbdsamples=10):
        super(Nbdsampler_multichannels, self).__init__()        
        self.nchannels = nchannels
        self.nnbdsamples = nnbdsamples
        self.comb_idx = list(it.product(range(self.nchannels),repeat=2))
        self.distance_cls = Distance_multichannels(nchannels)


    #def get_nbdindex(self, dist:Tensor, mask:Tensor):
    def get_nbdindex(self, dist:Tensor):        
        #dist[~mask[:, None, :].expand(*dist.shape)] = 1e8
        group_dist, group_idx = torch.topk(dist, self.nnbdsamples, dim=-1, largest=False, sorted=False)
        return group_dist,group_idx

        
    #def sample_nbdpoints(self,inputs: Tuple[Tensor, Tensor, Tensor], query_coords: Tensor, distance_param: Tensor):                         
    def sample_points(self,inputs: Tuple[Tensor, Tensor, Tensor, Tensor], query_coords: Tuple[Tensor,Tensor], distance_param: Tensor):                         
        
        x_a,ch_a, y_a, mask_a = inputs
        x_s,ch_s = query_coords

        #distance_selected = self.distance_cls.compute_distance((x_a,ch_a),(x_s,ch_s),distance_param)
        distance_selected = self.distance_cls.compute_distance((x_s,ch_s),(x_a,ch_a),distance_param)        
        neighbor_dist,neighbor_indices = self.get_nbdindex(distance_selected)

        neighbor_coords = index_points(x_a, neighbor_indices)
        neighbor_channels = index_points(ch_a, neighbor_indices)       
        neighbor_values = index_points(y_a, neighbor_indices)
        neighbor_mask = index_points(mask_a, neighbor_indices)
        #return neighbor_coords,neighbor_channels, neighbor_values, neighbor_mask
        return neighbor_dist,neighbor_coords,neighbor_channels, neighbor_values, neighbor_mask
    
        
        
```

```python
from typing import Tuple
import torch
import torch.nn as nn
from torch import Tensor

class Apply(nn.Module):
    """Support Module that apply input[dim] to the module
    """
    def __init__(self, module, dim=1):
        super().__init__()
        self.module = module
        self.dim = dim

    def forward(self, x):
        xs = list(x)
        xs[self.dim] = self.module(xs[self.dim])
        return xs

    def extra_repr(self):
        return "dim={}".format(self.dim)
    
    
class Swish(nn.Module):
    """Swish Module"""
    def __init__(self, inplace=False):
        super().__init__()

        self.inplace = inplace

    def forward(self, x):
        if self.inplace:
            x.mul_(torch.sigmoid(x))
            return x
        else:
            return x * torch.sigmoid(x)
        
        

class LinearBlock(nn.Module):
    """Linear -> BN -> Activation (Swish / ReLU)"""
    def __init__(
            self,
            in_features: int,
            out_features: int,
            activation: nn.Module = None,
            use_bn: bool = False
    ):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.linear = nn.Linear(in_features, out_features)
        self.bn = MaskBatchNormNd(out_features) if use_bn else None
        self.activation = Apply(Swish(inplace=True) if activation is None else nn.ReLU(inplace=True))

    def forward(self, x: Tensor):
        coords, values, mask = x
        numel = values.shape[:-1]
        values = self.linear(values.reshape(-1, self.in_features)).reshape(*numel, self.out_features)
        h = (coords, values, mask)
        if self.bn is not None:
            h = self.bn(h)
        h = self.activation(h)
        return h
    
    
    
    
class WeightNet(nn.Module):
    """Neural Network for Weight of Convolution
    Args:
        in_features (int): Number of input features
        out_features (int): Number of output features
        hidden_features (int): Number of hidden features
        activation (nn.Module, optional): activation
        use_bn (bool, optional): Whether the layer uses a batch normalization
    """
    def __init__(
            self,
            in_features: int,
            out_features: int,
            hidden_features: int = 32,
            activation: nn.Module = None,
            use_bn: bool = False,
    ):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.hidden_features = hidden_features
        self.use_bn = use_bn
        self.net = nn.Sequential(
            LinearBlock(in_features, hidden_features, activation, use_bn),
            LinearBlock(hidden_features, hidden_features, activation, use_bn),
            LinearBlock(hidden_features, out_features, activation, use_bn),
        )

    def forward(self, x: Tensor):
        return self.net(x)

    def __repr__(self):
        main_str = self._get_name() + '('

        main_str += 'in_features={}, out_features={}, hidden_features={}, bn={}'.format(
            self.in_features, self.out_features, self.hidden_features, self.use_bn
        )
        main_str += ')'
        return main_str

    
    
class PointConv(nn.Module):
    """Applies a point convolution over an input signal composed of several input points.
    Args:
        in_channels (int): Number of channels (features) in the input points
        out_channels (int): Number of channels (features) produced by the convolution
        mid_channels (int, optional): Number of channels (features) produced by the convolution
        num_nbhd (int, optional): Number of neighborhood points
        coords_dim (int, optional): Dimension of the input points
        activation (nn.Module, optional): activation for weight net
        use_bn (bool, optional): Whether the weight net uses a batch normalization
    """
    def __init__(
            self,
            in_channels: int,
            out_channels: int,
            mid_channels: int = 16,
            num_nbhd: int = 32,
            coords_dim: int = 3,
            sampling_fraction: float = 1,
            knn_channels: int = None,
            activation: nn.Module = None,
            use_bn: bool = False,
            mean: bool = False
    ):
        super().__init__()

        self.in_channels = in_channels
        self.out_channels = out_channels
        self.mid_channels = mid_channels
        self.num_nbhd = num_nbhd
        self.coords_dim = coords_dim
        self.sampling_fraction = sampling_fraction
        self.mean = mean

        self.weightnet = WeightNet(coords_dim, mid_channels, activation=activation, use_bn=use_bn)
        self.linear = nn.Linear(mid_channels * in_channels, out_channels)

    def forward(self, inputs: Tuple[Tensor, Tensor, Tensor]):
        """
        Args:
            inputs: [coords, values, mask], [(B, N, D), (B, N, C), (B, N)]
        """
        query_coords, query_values, query_mask = self.subsample(inputs)
        nbhd_coords, nbhd_values, nbhd_mask = self.extract_neighborhood(inputs, query_coords)
                
        # (B, M, nbhd, D) = (B, M, 1, D) - (B, M, nbhd, D)
        coords_deltas = self.get_embedded_group_elements(query_coords.unsqueeze(2), nbhd_coords)
        convolved_values = self.point_conv(coords_deltas, nbhd_values, nbhd_mask)
        convoleved_wzeros = torch.where(query_mask.unsqueeze(-1),
                                        convolved_values,
                                        torch.zeros_like(convolved_values))
        return query_coords, convoleved_wzeros, query_mask

    
    def extract_neighborhood(self,inputs: Tuple[Tensor, Tensor, Tensor], query_coords: Tensor):
        """Extract neighborhood
        Args:
            inputs: [coords, values, mask], [(B, N, D), (B, N, C), (B, N)]
            query_coords: (B, M, D)
        Returns:
            [(B, M, nbhd, D), (B, M, nbhd, C), (B, M, nbhd)]
            where nbhd = min(nbhd, N)
        """
        coords, values, mask = inputs
        neighbor_indices = knn_points(min(self.num_nbhd, coords.size(1)),
                                      coords[:, :, :self.knn_channels],
                                      query_coords[:, :, :self.knn_channels],
                                      mask)  # [B, M, nbhd]
        
        neighbor_coords = index_points(coords, neighbor_indices)
        neighbor_values = index_points(values, neighbor_indices)
        neighbor_mask = index_points(mask, neighbor_indices)
        return neighbor_coords, neighbor_values, neighbor_mask


    
    def get_embedded_group_elements(self, output_coords: Tensor, nbhd_coords: Tensor):
        return output_coords - nbhd_coords

    
    def point_conv(self, embedded_group_elements: Tensor, nbhd_values: Tensor, nbhd_mask: Tensor):
        """Operating point convolution
        Args:
            embedded_group_elements: (B, M, nbhd, D)
            nbhd_value: (B, M, nbhd, Cin)
            nbhd_mask: (B, M, nbhd)
        Returns:
            convolved_value (B, M, Cout)
        """
        B, M, nbhd, C = nbhd_values.shape
        _, penultimate_kernel_weights, _ = self.weightnet((None, embedded_group_elements, nbhd_mask))  # (B, M, nbhd, cmco_ci)
        masked_penultimate_kernel_weights = torch.where(nbhd_mask.unsqueeze(-1),penultimate_kernel_weights,torch.zeros_like(penultimate_kernel_weights).to(penultimate_kernel_weights.device))
        masked_nbhd_values = torch.where(nbhd_mask.unsqueeze(-1),nbhd_values,torch.zeros_like(nbhd_values).to(nbhd_mask.device))

        # (B, M, C_in, nbhd) x (B, M, nbhd, cmco_ci) => (B, M, C_in, cmco_ci)
        partial_convolved_values = masked_nbhd_values.transpose(-1, -2).matmul(masked_penultimate_kernel_weights).reshape(B, M, -1)
        convolved_values = self.linear(partial_convolved_values)  # (B, M, C_out)
        if self.mean:
            convolved_values /= nbhd_mask.sum(-1, keepdim=True).clamp(min=1)
        return convolved_values
    

    def extra_repr(self):
        line = 'C_in={}, C_out={}, C_mid={}, '.format(
            self.in_channels, self.out_channels, self.mid_channels
        )
        line += 'coords_dim={}, nbhd={}, sampling_fraction={}, mean={}'.format(
            self.coords_dim, self.num_nbhd, self.sampling_fraction, self.mean
        )
        return line
    
    
```

```python
from typing import Tuple
import torch
import torch.nn as nn
from torch import Tensor

class Apply(nn.Module):
    """Support Module that apply input[dim] to the module
    """
    def __init__(self, module, dim=1):
        super().__init__()
        self.module = module
        self.dim = dim

    def forward(self, x):
        xs = list(x)
        xs[self.dim] = self.module(xs[self.dim])
        return xs

    def extra_repr(self):
        return "dim={}".format(self.dim)
    
    
class Swish(nn.Module):
    """Swish Module"""
    def __init__(self, inplace=False):
        super().__init__()

        self.inplace = inplace

    def forward(self, x):
        if self.inplace:
            x.mul_(torch.sigmoid(x))
            return x
        else:
            return x * torch.sigmoid(x)
        
        

class LinearBlock(nn.Module):
    """Linear -> BN -> Activation (Swish / ReLU)"""
    def __init__(
            self,
            in_features: int,
            out_features: int,
            activation: nn.Module = None,
            use_bn: bool = False
    ):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.linear = nn.Linear(in_features, out_features)
        self.bn = MaskBatchNormNd(out_features) if use_bn else None
        self.activation = Apply(Swish(inplace=True) if activation is None else nn.ReLU(inplace=True))

    def forward(self, x: Tensor):
        coords, values, mask = x
        numel = values.shape[:-1]
        values = self.linear(values.reshape(-1, self.in_features)).reshape(*numel, self.out_features)
        h = (coords, values, mask)
        if self.bn is not None:
            h = self.bn(h)
        h = self.activation(h)
        return h
    
    
    
    
class WeightNet(nn.Module):
    """Neural Network for Weight of Convolution
    Args:
        in_features (int): Number of input features
        out_features (int): Number of output features
        hidden_features (int): Number of hidden features
        activation (nn.Module, optional): activation
        use_bn (bool, optional): Whether the layer uses a batch normalization
    """
    def __init__(
            self,
            in_features: int,
            out_features: int,
            hidden_features: int = 16,
            activation: nn.Module = None,
            use_bn: bool = False,
    ):
        super().__init__()
        self.in_features = in_features
        self.out_features = out_features
        self.hidden_features = hidden_features
        self.use_bn = use_bn
        self.net = nn.Sequential(
            LinearBlock(in_features, hidden_features, activation, use_bn),
            LinearBlock(hidden_features, hidden_features, activation, use_bn),
            LinearBlock(hidden_features, out_features, activation, use_bn),
        )

    def forward(self, x: Tensor):
        return self.net(x)

    def __repr__(self):
        main_str = self._get_name() + '('

        main_str += 'in_features={}, out_features={}, hidden_features={}, bn={}'.format(
            self.in_features, self.out_features, self.hidden_features, self.use_bn
        )
        main_str += ')'
        return main_str

    
    
class PointConv_multichannels(nn.Module):
    """Applies a point convolution over an input signal composed of several input points.
    Args:
        in_channels (int): Number of channels (features) in the input points
        out_channels (int): Number of channels (features) produced by the convolution
        mid_channels (int, optional): Number of channels (features) produced by the convolution
        num_nbhd (int, optional): Number of neighborhood points
        coords_dim (int, optional): Dimension of the input points
        activation (nn.Module, optional): activation for weight net
        use_bn (bool, optional): Whether the weight net uses a batch normalization
    """
    def __init__(
            self,
            in_channels: int,
            out_channels: int,
            mid_channels: int = 16,
            num_nbd: int = 32,
            pos_dim: int = 3,
            num_channels: int=3,
            sampling_fraction: float = 1,
            activation: nn.Module = None,
            use_bn: bool = False,
    ):
        super().__init__()

        self.pos_dim = pos_dim
        self.num_channels = num_channels

        self.in_channels = in_channels
        self.out_channels = out_channels
        self.mid_channels = mid_channels
        self.num_nbd = num_nbd        
        
        self.sampling_fraction = sampling_fraction                
        self.nbd_sampler = Nbdsampler_multichannels(nchannels=self.num_channels, nnbdsamples=self.num_nbd)
        self.weight_pointconv = WeightNet(in_features=1, out_features=self.in_channels, hidden_features = self.mid_channels, activation=activation, use_bn=use_bn) # 
        #self.linear = nn.Linear(mid_channels * in_channels, out_channels)
        self.linear = nn.Linear(self.in_channels, self.out_channels)


    #def forward(self, inputs: Tuple[Tensor,Tensor, Tensor, Tensor],comb_param: Tensor):        
    #def forward(self, xc_vec, xc_ch, yc_vec, xc_bool, comb_param):        
    def forward(self, inputs: Tuple[Tensor,Tensor, Tensor, Tensor,Tensor]):        
        
        """
        Args:
            inputs: [coords,coords_ch, values, mask], [(B, N, D), (B, N, C), (B, N)]
        """
        xc_vec,xc_ch,yc_vec,xc_bool,comb_param = inputs
        nbd_dist,nbd_xc,nbd_ch,nbd_yc,nbd_bool = self.nbd_sampler.sample_points((xc_vec,xc_ch,yc_vec,xc_bool),(xc_vec,xc_ch),comb_param) 
        conv_weights = self.weight_pointconv((None,nbd_dist.unsqueeze(dim=-1).detach().clone(),None))[1]
        convolved_out = (conv_weights*nbd_yc).mean(dim=-2)
        return xc_vec,xc_ch,self.linear(convolved_out),xc_bool,comb_param            


    def extra_repr(self):
        line = 'C_in={}, C_out={}, C_mid={}, '.format(self.in_channels, self.out_channels, self.mid_channels)
        #line += 'coords_dim={}, nbhd={}, sampling_fraction={}, mean={}'.format(self.coords_dim, self.num_nbhd, self.sampling_fraction, self.mean)
        return line


```

```python

#def vector_repr_1d(x_c,y_c,tovec=True):
def origin_to_vector1d(x_c,y_c):    
    assert len(x_c.size())==3
    nb,nseqlen,nchxndim = x_c.size()
    out,out2 =[],[]
    for j in range(x_c.size(dim=-1)):
        tmp_ch = x_c[:,:,j].unsqueeze(dim=-1)
        tmp_ch = torch.cat([tmp_ch,(j*torch.ones_like(tmp_ch)).to(tmp_ch.device)],dim=-1)
        out.append(tmp_ch)
        tmp_ch2 = y_c[:,:,j].unsqueeze(dim=-1)
        out2.append(tmp_ch2)
    out = torch.cat(out,dim=-2)
    out2 = torch.cat(out2,dim=-2)    
    assert out.size(-1) == 2

    out_pos,out_ch = out.split((out.size(-1)-1,1),dim=-1)
    out_val = out2 
    out_bool = torch.ones_like(out_pos).to(out_pos.device).bool()    
    return out_pos,out_ch.int()[...,0],out_val,out_bool

    
def vector1d_to_origin(xc_v,xc_ch,pmustd_v,nxdim=1,nydim=1):
    nb = xc_v.size(0)    
    outs_x,outs_y,outs_y2=[],[],[]
    for j in range(3):
        outs_x.append(xc_v[xc_ch == j].reshape(nb,-1,nxdim))
        outs_y.append(pmustd_v[0][xc_ch == j].reshape(nb,-1,nydim))
        outs_y2.append(pmustd_v[1][xc_ch == j].reshape(nb,-1,nydim))

    outs_x=torch.cat(outs_x,dim=-1)
    outs_y=torch.cat(outs_y,dim=-1)
    outs_y2=torch.cat(outs_y2,dim=-1)    
    return outs_x.to(xc_v.device),outs_y.to(xc_v.device),outs_y2.to(xc_v.device)




class Pconvcnp_multichannel(nn.Module):
    def __init__(self,in_dim: int = 1,out_dim:int = 1, num_channels: int=3, num_nbd: int=10):
        super(Pconvcnp_multichannel,self).__init__()
        
        self.num_nbd = num_nbd
        
        pconv1 = PointConv_multichannels(in_channels=1,out_channels=10,num_nbd=self.num_nbd, pos_dim= 1,num_channels=3)
        pconv2 = PointConv_multichannels(in_channels=10,out_channels=10,num_nbd=self.num_nbd, pos_dim= 1,num_channels=3)
        pconv3 = PointConv_multichannels(in_channels=10,out_channels=2*1,num_nbd=self.num_nbd, pos_dim= 1,num_channels=3)
        
        self.pconv_all = nn.Sequential(pconv1,
                                       Apply(nn.ReLU(), dim=1),
                                       pconv2,
                                       Apply(nn.ReLU(), dim=1),                                       
                                       pconv3)
        
        self.pred_outs = Conv_deepset(in_dims=1,out_dims=1)
        
        
        
    def forward(self,xc,yc,xt,yt=None):        
        xc_v,xc_ch,yc_v,xc_bool=origin_to_vector1d(xc,yc)
        
        # samples the sample path by GP.
        """
        
        """
        
        xc_v1,xc_ch1,yc_v1,xc_bool1,_ = self.pconv_all( (xc_v,xc_ch,yc_v,xc_bool, comb_paraml) )
        pmu_yc_v,pstd_yc_v = yc_v1.split((1,1),dim=-1)    
        xct,pmu_yct,pstd_yct = vector1d_to_origin(xc_v1,xc_ch,(pmu_yc_v,pstd_yc_v))
        
        # interpolations by kernel smmothing 
        """
        
        """
        
        return xct,pmu_yct,pstd_yct
        
```

```python
class Conv_deepset(nn.Module):
    #def __init__(self,in_channels=1,out_channles=1):
    def __init__(self,in_dims=1,out_dims=1):
        
        super(Conv_deepset,self).__init__()
        
        #self.in_channels = in_channels
        #self.out_channels = out_channels
        

        self.in_dims = in_dims
        self.out_dims = out_dims
        self.scale_parameters = None
        
        
        
    def compute_rbf(self,x1,x2=None):
        if x2 is None:
            x2 = x1            
        # Compute shapes.            
        nbatch,npoints,nchannel = x1.size()
        
        #compute rbf over multiple channels
        dists = x1.unsqueeze(dim=2) - x2.unsqueeze(dim=1)        
        #scales = self.sigma_fn(self.sigma)[None, None, None, :]                
        
        #factors = 1
        #if dists.size(-1) != scales.size(-1):
        #    factors = scales.size(-1) // dists.size(-1) 
        #    dists = dists.repeat(1,1,1,factors)        
        #dists /= (scales + eps)
        
        wt = torch.exp(-0.5*dists**2)   
        return wt

    
    def forward(self,x_ct,pmustd_ct,x_t):
        nbatch,npoints,nchannel = x_t.size()
        wt = self.compute_rbf(x_ct,x_t)
        h0 = wt.sum(dim=1)

        pmu_ct,pstd_ct = pmustd_ct
        print('pmu_ct.size(),pstd_ct.size(),wt.size()')
        print(pmu_ct.size(),pstd_ct.size(),wt.size())        
        
        pmu_t = (pmu_ct.unsqueeze(dim=-2)*wt).sum(dim=1)
        pstd_t = (pstd_ct.unsqueeze(dim=-2)*wt).sum(dim=1)
        
        return pmu_t,pstd_t
        #return y_out,h1,h0
        #return y_out,n_h1,h1,h0

    
    
    def extra_repr(self):
        line = 'C_in={}, C_out={}, '.format(self.in_dims, self.out_dims)
        #line += 'coords_dim={}, nbhd={}, sampling_fraction={}, mean={}'.format(self.coords_dim, self.num_nbhd, self.sampling_fraction, self.mean)
        return line
    
            
        
        
        
        
        
```
